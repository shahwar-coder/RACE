## **7. Global Variables**

```python
x = "awesome"

def myfunc():
  print("Python is " + x)

myfunc()

# This is just a simple scenario, where x is assigned outside and used (can be used in function as well)
```

```python
x = "awesome"

def myfunc():
  x = "fantastic"
  print("Python is " + x)

myfunc()

print("Python is " + x)

# In this scenario first outer x is assigned "awesome"
# The we come across a function, but we do not go inside since have not encountered a call to the function yet.
# As we move forward we see a call to function, now we have x now pointing to another value which is "fantastic"
# Point to note is there is no new x made, only the original x now pointing to new string value.
# The function also has a print for x, so we see fantastic printed
# Now as we move, we now encounter a fresh outside print (the last statement)
# Again x will now point to "awesome" as it initially did. When it pointed to "fantastic" it was local scope.
# Finally it prints "awesome"
```

```python
x = "awesome"

def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)

# This is same code a before, only we add global keyword infront of local x inside the function
# What this essentially does is, it has now the power to override the outside x, x taking any affect inside the function will now be refleted outside the function as well.
```

```python
# If you use the global keyword, the variable belongs to the global scope:
def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)

# Thsi is another testing with no initial global declaration of x
```
